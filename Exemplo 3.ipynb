```python
# espera por um serie de valores.
reviews_media = reviews['Quantidade Vendida'].mean()
reviews['Quantidade Vendida'].map(lambda p: p - reviews_media)
```

```python
def remean_points(row):
    row['Quantidade Vendida'] = row['Quantidade Vendida'] - reviews_media
    return row

reviews.apply(remean_points, axis='columns')
```
