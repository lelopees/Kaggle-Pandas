```python
import pandas as pd
```

```python
reviews = pd.read_csv(r'C:\Users\leticia.lopes\Desktop\Estudos\Python\Didática Tech\Pandas\athlete_events.csv')
```

```python
reviews.head()
```

```python
reviews['Year'].value_counts()
```

```python
# sumarizar o agrupamento pelos campos de Year e Noc
# agrupamentos por mais de um campo
# reset_index deixa os indices 
reviews.groupby(['Year', 'NOC']).sum().reset_index()
```

```python
# agg metódo agregador para Age será aplicado minino, para Height será aplicado max
reviews.groupby(['Year', 'NOC']).agg({'Age': 'min', 'Height': 'max'})
```

```python
reviews.groupby('Year').Year.min()
```

```python
reviews.groupby('City')[''
```

```python
# obter a primeira cidade de cada Team
reviews.groupby('Team').apply(lambda df: df['City'].iloc[0])
```

```python
# obter a maior idade por season e sport
reviews.groupby(['Season','Sport']).apply(lambda df: df.loc[df.Age.idxmax()])
```

```python
# obter várias função da idade len,min e max por esporte
reviews.groupby(['Sport']).Age.agg([len,min,max])
```

```python
sporties_df = reviews.groupby(['Sport','City']).Age.agg([min])
display(sporties_df)
```

```python
mi = sporties_df.index
type(mi)
```

```python
# para um indice normal
sporties_df = sporties_df.reset_index()
sporties_df
```

```python
# por padrão os numeros mais baixos vão primeiro
sporties_df.sort_values(by='min')
```

```python
# para trazer os valores mais altos primeiro
sporties_df.sort_values(by='min', ascending=False)
```

```python
# classificar os valores de indices
sporties_df.sort_index()
```

```python
# classificar por mais de uma coluna por vez
sporties_df.sort_values(by=['City', 'min'])
```
